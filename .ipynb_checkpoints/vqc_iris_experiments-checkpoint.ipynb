# Variational Quantum Classifier - Parameters experimentation

_____________


In this notebook, we will expore different architectures and hyperparameters for our VQC. For more details on the model, see [this notebook.](./vqc_iris_fundamentals.ipynb)

_____________

**Import data analysis/visualization libraries**

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import classification_report, confusion_matrix
```

**Iterative and time tools for the parameters experimentation**

```python
import itertools

import time
```

**Reading the data**

```python
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

**Preprocess**

```python
# map between integer label and categorical target

cat_map = dict(zip(df["species"].astype("category").cat.codes, df["species"]))
print("\nTarget integer-label encoding:", cat_map, "\n\n")

# categorical target --> integer label

df["species"] = df["species"].astype("category").cat.codes

#################################################################################

# taking only species 1 and 0 -- binary classification

df = df[df["species"].isin([0, 1])]

#################################################################################

df.info()
```

```python
# features
X = df.drop(columns="species")

# target
y = df["species"]
```

```python
# train-test split, 20% test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=42, 
                                                    stratify=y)
```

___________


## Variational Quantum Classifier

Let's import the .py file with the VQC functions!

```python
from vqc import *
```

___________


__Grid search of parameters__

Let's search over combinations of learning rate, loss function, feature maps and number of layers (repetitions).

```python
# for reproducibility
np.random.seed(42)

N = X_train.shape[1]

# initial theta
theta = np.random.uniform(0, 2*np.pi, N)

# parameters
num_epochs = 20
n_runs = 1e3

# params to be varied in the grid
params = {"lr_list" : [1e-3, 1e-2, 5e-2, 1e-1],
          "loss_func_list" : ["BCE", "AE", "SE"],
          "feature_map_list" : ["ZZ", "amplitude"],
          "rep_fm_list" : [1, 2],
          "rep_var_list" : [1, 2]}

# combination of parameters
param_list = list(itertools.product(params["lr_list"],
                                    params["loss_func_list"],
                                    params["feature_map_list"],
                                    params["rep_fm_list"],
                                    params["rep_var_list"]))

# dictionary with param grid above
param_grid = {key.replace("_list", "") : [x[i] for x in param_list] for i, key in enumerate(params.keys())}

# dictionary to save results
grid_results = {key : [] for key in param_grid.keys()}
grid_results["final_theta"] = []
grid_results["mean_training_accuracy"] = []
grid_results["weighted_test_f1"] = []
grid_results["test_accuracy"] = []
```

```python
for i in range(len(param_list)):
    
    start = time.time()
    
    feature_map = param_grid["feature_map"][i] if "feature_map" in param_grid else "amplitude"
    lr = param_grid["lr"][i] if "lr" in param_grid else 5e-2
    loss_func = param_grid["loss_func"][i] if "loss_func" in param_grid else "SE"
    rep_fm = param_grid["rep_fm"][i] if "rep_fm" in param_grid else 1
    rep_var = param_grid["rep_var"][i] if "rep_var" in param_grid else 1
    
    print("\n")
    print("="*90)
    print("Combination {}/{}:".format(i+1, len(param_list)))
    print("="*90)
    print("\t \t \t Feature map: {}".format(feature_map))
    print("\t \t \t Reps FM: {}".format(rep_fm))
    print("\t \t \t Reps var layer: {}".format(rep_var))
    print("\t \t \t Learning rate: {}".format(lr))
    print("\t \t \t Loss function: {}".format(loss_func))
    print("="*90)
    print("\n")
    
    training_loss, training_accuracy, theta_final = stochastic_gd(N, X_train, y_train, theta=theta,
                                                                  loss_func=loss_func, lr=lr, delta=5e-3,
                                                                  num_epochs=num_epochs, seed_simulator=42,
                                                                  n_runs=n_runs, feature_map=feature_map,
                                                                  rep_fm=rep_fm, rep_var=rep_var)
    
    #############################################################
    #############################################################
    #############################################################
    
    plt.title("Loss function during training")
    
    sns.lineplot(x=range(len(training_loss)), y=training_loss, color="blue")
    
    plt.xlabel("Epoch #")
    plt.ylabel("Loss")
    
    plt.show()
    
    ##############################
    
    plt.title("Accuracy during training")
    
    sns.lineplot(x=range(len(training_accuracy)), y=training_accuracy, color="orange")
    
    plt.xlabel("Epoch #")
    plt.ylabel("Accuracy")
    
    plt.show()
    
    #############################################################
    #############################################################
    #############################################################
    
    print("Final variational parameters vector:", theta_final)
    
    mean_acc_train = np.mean(training_accuracy)
    print("\nMean training accuracy in the {} pochs: {}".format(num_epochs, mean_acc_train))
    
    #############################################################
    #############################################################
    #############################################################
    
    # predictions on test data
    
    y_pred = []
    
    for k in range(X_test.shape[0]):
        
        qc = build_circuit(N, 1, X_test, theta_final, k, show_circuit=False, 
                           feature_map=feature_map, rep_fm=rep_fm, rep_var=rep_var)

        _, y_pred_k = vqc(qc, seed_simulator=42)

        y_pred.append(y_pred_k)
        
        
    print("Quantum classifier (VQC):\n")

    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    
    cr_dict = classification_report(y_test, y_pred, output_dict=True)
    f1_test = cr_dict["weighted avg"]["f1-score"]
    acc_test = cr_dict["accuracy"]
    
    if "feature_map" in param_grid: grid_results["feature_map"].append(feature_map)
    if "lr" in param_grid: grid_results["lr"].append(lr)
    if "loss_func" in param_grid: grid_results["loss_func"].append(loss_func) 
    if "rep_fm" in param_grid: grid_results["rep_fm"].append(rep_fm)
    if "rep_var" in param_grid: grid_results["rep_var"].append(rep_var)
        
    grid_results["final_theta"].append(theta_final)
    grid_results["mean_training_accuracy"].append(mean_acc_train)
    grid_results["weighted_test_f1"].append(f1_test)
    grid_results["test_accuracy"].append(acc_test)
    
    #############################################################
    #############################################################
    #############################################################
    
    stop = time.time()
    duration = stop-start

    print("\n\n---------> Training duration: {}".format(time.strftime("%H:%M:%S", time.gmtime(duration))))
    
print("\nGrid results:\n")

evaluation = pd.DataFrame(grid_results)

evaluation["num_epochs"] = num_epochs
evaluation["n_runs"] = n_runs

evaluation = evaluation.sort_values(["mean_training_accuracy",
                                     "test_accuracy",
                                     "weighted_test_f1"], ascending=False)

display(evaluation)
```

Great! Some architectures achieved perfect performance! 

Let's see the top 5:

```python
evaluation.head(5)
```

The amplitude feature map with 2 repeated variational layers yielded best results, that's interesting!

So, indeed, there are better choices when it comes to VQC's architecutre, and a grid search-like procedure can be employed to determine  parameters leading to very good models, even with this very simple implementation.

Notice, however, that we stick to very simple variational layers and feature maps, since our main goal was to understand how the model works. Further exploring the space of potential feature maps/variational layers may lead to better models when trained on more complicated datasets. This will be done in future repositories, but probably employing other implementations (those of PennyLane or Qiskit, for instance), which may make it easier to experiment over possible choices.
